```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.charts import Bar
from bokeh.palettes import inferno,plasma, viridis
from bokeh.models import Legend
```

```python
data = pd.read_csv("C:\\Users\\fatemeh\\Desktop\\Hackaton\\data_breast_cancer_less_50.csv", skiprows = 5)
```

```python
data.head()
data.info()
data.isnull().any()
```

```python
data = data.dropna()
data.isnull().any()
data.info()
```

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select,lasso_select"
output_notebook()

x_1 = data.loc[data['Race/Ethnicity'] == 'White', 'Year']
y_1 = data.loc[data['Race/Ethnicity'] == 'White', 'Rate']

x_2 = data.loc[data['Race/Ethnicity'] == 'Black', 'Year']
y_2 = data.loc[data['Race/Ethnicity'] == 'Black', 'Rate']

fig = figure( width =900, height= 400,  tools= TOOLS, title= 'Age-Adjusted SEER Incidence Rates, By Race and Sex, For Breast Cancer In Females  Ages < 50, During 1975-2015  (SEER 9)')

fig.line(x_1, y_1,  line_color = 'blue', legend = 'White Female')
fig.circle(x_1, y_1, fill_color="blue", size=8)

fig.line(x_2, y_2, line_color = 'red', )
fig.square(x_2, y_2, fill_color="red", size=8, legend = 'Black Female')

fig.xaxis.axis_label = 'Year of Diagnosis'
fig.yaxis.axis_label = 'Rate per 100,000'

show(fig)
```
